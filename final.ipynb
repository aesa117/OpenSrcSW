```python
import csv
f = open('country_vaccinations.csv')
data = csv.reader(f)
next(data)

max_val = -99
max_day = ""
min_val = 99
min_day = ""

for row in data:
    if row[7] != '':
        row[7] = float(row[7])
        if row[7] > max_val:
            max_val = row[7]
            max_day = row[2]
        if row[7] < min_val:
            min_val = row[7]
            min_day = row[2]
            
print("백신이 가장 많이 접종된 날 {0:s}, 가장 적게 접종된 날 {1:s}".format(max_day, min_day))
```
